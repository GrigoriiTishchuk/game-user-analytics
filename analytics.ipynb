Preparing the data for retention check

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("game_data.csv")
df["event_date"] = pd.to_datetime(df["event_date"])
df["install_date"] = pd.to_datetime(df["install_date"])
df["days_since_install"] = (df["event_date"] - df["install_date"]).dt.days
```

Building user activity table

```python
user_days = df.groupby("user_id")["days_since_install"].apply(set)
```

Computing retention


```python
total_users = len(user_days)

day1_retention = sum(1 for days in user_days if 1 in days) / total_users
day7_retention = sum(1 for days in user_days if 7 in days) / total_users
```
